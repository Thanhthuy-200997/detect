```python
import os
import glob
import pandas as pd
import numpy as np
import math

def train_val_test_split(lst_data:list):
    idx1 = int(round(len(lst_data)*0.8,0))
    train = lst_data[:idx1]
    sub = lst_data[idx1:]
    idx2 = int(round(len(sub)*0.5,0))
    val = sub[:idx2]
    test = sub[idx2:]
    return train,val,test
```

```python
import cv2
from torchvision.transforms import transforms
import torch



class Transform:
   def __init__(self):
      super(Transform,self).__init__()
      # self.tf = transforms.ToTensor()
      self.tf = transforms.Compose([
        transforms.Resize((640,640)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                             std=[0.229, 0.224, 0.225])
        ])
      
   def transform(self,img):
      inputs = self.tf(img)
      input = torch.unsqueeze(inputs,0)
      return input
   
```

```python
from PIL import Image
from torchvision import transforms

tfms = transforms.Compose([transforms.Resize((224,224)), transforms.CenterCrop((224,224)), 
                           transforms.ToTensor(),
                           transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),])
```

```python
import glob
lst = glob.glob('D:/hh/detect/animals/*')
lst_vector = []
for i in lst:
    im = Image.open(i).convert('RGB')
    img = tfms(im).unsqueeze(0)
    with torch.no_grad():
        out = model(img)
        out = out.detach().numpy()
        out = out.flatten()
        print('-----------')
        print(out.shape)
        lst_vector.append(out)
    im.close()
    
    
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
label = [0,1,0]
df = pd.DataFrame(columns = ['vectors','labels'])
df['vectors'] = lst_vector
df['labels'] = label
X_train, X_test, y_train, y_test = train_test_split(df['vectors'], df['labels'], test_size = 0.2, random_state = 0)
#input data 
x=np.array(X_train.tolist())
#output data
y=np.array(y_train.tolist())
```

```python
from sklearn import svm
svc=svm.SVC(C = 1, kernel = 'rbf', probability=True)
svc.fit(x,y)
```

```python
from sklearn.metrics import accuracy_score
xt=np.array(X_test.tolist())
#output data
yt=np.array(y_test.tolist())
y_pred=svc.predict(xt)
print(y_pred)
print('Model accuracy score with linear kernel and C=100.0 : {0:0.4f}'. format(accuracy_score(yt, y_pred)))
```
